```python
import pandas as pd
```

```python
apple_data = pd.read_csv("../data/applemobilitytrends-2020-06-11.csv")
```

```python
apple_data.describe()
```

```python
apple_data.head()
```

```python
apple_data["alternative_name"].fillna("", inplace=True)
apple_data["sub-region"].fillna("", inplace=True)
apple_data["country"].fillna("", inplace=True)
```

```python
india_data = apple_data[apple_data["region"] == "India"]
```

```python
india_cities_data = apple_data[apple_data["country"].str.contains("India")]
```

```python
india_cities_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
```

```python
india_data.head()
```

```python
days = [col for col in india_data.columns if col.startswith("2020")]
days.append("transportation_type")
```

```python
xtick_labels = [d if (i % 14 == 0) and d.startswith("2020") else "" for i, d in enumerate(days)]
```

```python
plot_data = india_data[days].melt(id_vars="transportation_type")
sns.lineplot(data=plot_data, x="variable", y="value", hue="transportation_type")
plt.xticks(ticks=plot_data["variable"].unique(), labels=xtick_labels, rotation=45);
```

```python
days.append("region")
```

```python
plot_data = india_cities_data[days].melt(id_vars=["transportation_type", "region"])
sns.lineplot(data=plot_data, x="variable", y="value", hue="region", style="transportation_type")
plt.xticks(ticks=plot_data["variable"].unique(), labels=xtick_labels, rotation=45);
```

```python
import altair as alt
```

```python
alt.Chart(plot_data).mark_line().encode(x="variable:T", y="value", color="region", strokeDash="transportation_type")
```

```python
india_cities_data.head()
```

```python
plot_data.sort_values(["region", "transportation_type"], inplace=True)
plot_data.reset_index(drop=True, inplace=True)
```

```python
days.pop(-2)
```

```python
india_cities_data_walking = india_cities_data[india_cities_data["transportation_type"] == "walking"][days].melt(id_vars=["region"], value_name="walking", var_name="date")
```

```python
india_cities_data_driving = india_cities_data[india_cities_data["transportation_type"] == "driving"][days].melt(id_vars=["region"], value_name="driving", var_name="date")
```

```python
india_cities_data_driving.head()
```

```python
plot_data = india_cities_data_driving.merge(india_cities_data_walking)
```

```python
alt.Chart(plot_data).mark_line().encode(x="date:T", y="driving", color="region")
```

```python
plot_data.sort_values(["region", "date"], inplace=True)
plot_data.reset_index(drop=True, inplace=True)
```

```python
plot_data.head()
```

```python
plot_data["driving_rolling_3"] = plot_data.groupby("region")["driving"].rolling(3).mean().reset_index(drop=True)
```

```python
alt.Chart(plot_data).mark_line().encode(x="date:T", y="driving_rolling_3", color="region")
```
